```python
folders = [i for i in range(1, 8)]
for folder_num in folders:

    images = glob.glob("/home/alex/Turn_lights/rgb/train/%05i/*.png" %folder_num)
    images.sort()

    annotations = glob.glob("/home/alex/Turn_lights/annotations_bosch/%05i/*.txt" %folder_num)
    annotations.sort()

    save_path = "/home/alex/Turn_lights/dataset_bosch/%05i/" %folder_num
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    for i, im_path in enumerate(images):
        #print i
        #print im_path
        im = cv2.imread(im_path)
        df = pd.read_csv(annotations[i], sep=' ', header=None)
    
        is_car = df[0] == "Car"
        #is_big_car = df[0] == "BigCar"
        prob_more_08 = df[15] > 0.9
        df = df[is_car & prob_more_08]
        df[[4,5,6,7]] = df[[4,5,6,7]].astype(int)
    
        j = 0
        for line in df.iterrows():
            resized = cv2.resize(im[line[1][5]:line[1][7],line[1][4]:line[1][6]],
                                 (200,200), interpolation = cv2.INTER_AREA)
            #print save_path +"image_%04i_%2i.jpg" %(i, j)
            cv2.imwrite(save_path +"image_%04i_%02i.jpg" %(i, j), resized)
            j += 1
```
